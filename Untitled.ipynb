```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse

from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity, cosine_distances
```

```python
poetry_1 = pd.read_json('./Dataset/goodreads_books_poetry.json', lines=True)

poetry_1.drop(columns=['isbn', 'series', 'kindle_asin', 'asin', 'is_ebook', 'edition_information', 'publication_year', 'url', 'image_url', 'format', 'link', 'publisher', 'publication_day', 'isbn13', 'publication_month', 'work_id'], inplace=True)

poetry_1.head()
```

```python
for index in poetry_1.index:
    author_id = poetry_1['authors'][index][0]['author_id']
    poetry_1['authors'][index] = author_id

poetry_1.head()
```

```python
#Review dataframe: 
poetry_rev = pd.read_json('./Dataset/goodreads_reviews_poetry.json', lines=True)
poetry_rev.drop(columns=['date_added', 'date_updated', 'read_at', 'started_at', 'n_votes', 'n_comments'], inplace=True)

poetry_rev.head()
```

```python
#Create pivot table

data = poetry_rev.pivot_table(index = 'book_id', values = 'rating', columns = 'user_id')
data.head()
```

```python
sparse_data = sparse.csr_matrix(data.fillna(0))
```

```python
import sys

sys.getsizeof(data), sys.getsizeof(sparse_data)
```

## Cosine similarity

```python
cosine_distances(sparse_data)
```

```python
pairwise_distances(sparse_data, metric='cosine')
```

```python
similarity_matrix = cosine_similarity(sparse_data)
```

```python
sim_sparse = cosine_similarity(sparse_data, dense_output = False)
```

```python
pd.DataFrame(
    similarity_matrix,
    index = data.index,
    columns = data.index
)
```

```python
rec_df = pd.DataFrame.sparse.from_spmatrix(
    sim_sparse,
    index = data.index,
    columns = data.index
)
```
