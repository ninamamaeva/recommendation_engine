```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import sparse

from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity, cosine_distances
from sklearn.neighbors import NearestNeighbors
```

```python
def file_read_authors():
    df = pd.read_json('./Dataset/goodreads_book_authors.json', lines=True)
    df.rename(columns={'average_rating': 'author_avg_rating', 'author_id': 'authors', 'text_reviews_count': 'aut_txt_rev_count','name': 'author_name', 'ratings_count': 'author_ratings_count'}, inplace=True)
    return df


def file_read(genre):
    df = pd.read_json('./Dataset/goodreads_books_'+ genre +'.json', lines=True)
    df.drop(columns=['isbn', 'series', 'kindle_asin', 'popular_shelves','similar_books','asin', 'is_ebook', 'edition_information', 'publication_year', 'url', 'image_url', 'format', 'link', 'publisher', 'publication_day', 'isbn13', 'publication_month', 'work_id'], inplace=True)
    return df

def fix_author(df):
    '''Extracts the author id from the disctionary of authors'''
    for index in df.index:
        author_id = df['authors'][index][0]['author_id']
        df['authors'][index] = int(author_id)

# add the authors names to df:
def add_author_name(df, aut):
    return df.merge(aut[['author_name','authors']], on = 'authors', how = 'left')
```

```python
# program starts here

# init authors df
df_authors = file_read_authors()

# init individual genre dfs
df_poetry = file_read('poetry')
df_poetry['genre'] = 'poetry'

# df_children = file_read('children')
# df_poetry['children'] = 'children'

dfs = [df_poetry]

# process each df
for d in dfs:
    fix_author(d)

df_poetry = add_author_name(df_poetry, df_authors)

df_poetry.head(3)
    
```

```python
def plot_df(dfs):
    '''Plots the Language distribution data'''
    rows = 2
    cols = 2
    
    f, axs = plt.subplots(nrows = rows, ncols = cols, figsize = (20, 10))
    plt.subplots_adjust(hspace = 0.2, wspace = 0.2)
    
    count = 0
    for r in range(rows):
        for c in range(cols):
            kwargs = {'alpha':0.5}
            sns.countplot(x = dfs[count]['language_code'], data = dfs[count], ax = axs[r,c], **kwargs)
            axs[r,c].set_title('Language distribution for ' + dfs[count]['genre'].iloc[0])
            count += 1
            # axs[r,c].set_xtickslabels(dfs, rotation = 60) set x-ticks labesls
    plt.show()
```

```python
#Top books:
# [genre1, genre2, etc], each one is a df
def plot_top_books(dfs):
    all_top_books = []

    for df in dfs:
        top_books = df[df['ratings_count'] > 100000]
        top_books = top_books.sort_values(by='average_rating', ascending=False).head(20)
        all_top_books.append(top_books)

    '''Plots the top books'''
    rows = 2
    cols = 2
    
    f, axs = plt.subplots(nrows = rows, ncols = cols, figsize = (20, 10))
    plt.subplots_adjust(hspace = 0.2, wspace = 0.2)
    
    count = 0
    for r in range(rows):
        for c in range(cols):
            kwargs = {'alpha':0.5}
            sns.barplot(x = all_top_books[count]["average_rating"], y = all_top_books[count]['title'],data = all_top_books[count], ax = axs[r,c], **kwargs)
            axs[r,c].set_title('top books for ' + dfs[count]['genre'].iloc[0])
            count += 1
            # axs[r,c].set_xtickslabels(dfs, rotation = 60) set x-ticks labesls
    plt.show()   

plot_top_books(dfs_1) 
```
